```python
import tweepy


consumer_key="xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
consumer_secret="xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
access_key="xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
access_secret="xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"



auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)
api=tweepy.API(auth)
```

```python
#sudhir chowdary
#ravish kumar
#nitin gadkari
#burkhadutt
#arvind kejriwal
#shashi tharoor

import operator

followers_data=[]
foll=tweepy.Cursor(api.followers, '@ShashiTharoor').items()

for i in foll:
    print(len(followers_data))
    followers_data.append(i._json)

```

```python
top_followers={}

for i in followers_data:
    top_followers[i['screen_name']]=i['followers_count']

sorted_followers = dict( sorted(top_followers.items(), key=operator.itemgetter(1),reverse=True))
list_of_followers=list(sorted_followers.keys())[:10]
print(list_of_followers)
```

```python


p=api.user_timeline(screen_name = 'TejpalTamu',tweet_mode='extended',count=200)
data=list(p)
old_id=data[-1]._json['id']-1
while ((len(p)!=0) and (len(data)<500)):
    p=api.user_timeline(screen_name = 'TejpalTamu',tweet_mode='extended',count=200,max_id=old_id)
    data.extend(list(p))
    old_id=data[-1]._json['id']-1
    
```

```python
import csv
q=0
with open('SudhirChaudary.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for i in data:
        try :
            writer.writerow([i._json['full_text']])
        except Exception as e :
            q+=1
            print(q)
```
